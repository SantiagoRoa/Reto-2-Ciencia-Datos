```
# CREACION DE DATA FRAME #

df = pd.DataFrame.from_dict(tweets)
df.head(100)
```

```
# PALABRAS CLAVES DE BUSQUEDA #

petro_wordKeys = ['Gustavo Petro', 'GustavoPetro', 'Petro', 'Gustavo', 'petrogustavo', 'Petro\'s']
zuluaga_wordKeys = ['Oscar Ivan Zuluaga', 'OIZuluaga', 'Zuluaga', 'OscarIvan', 'IvanZuluaga', 'Ivan','Oscar','Oscar Zuluaga','Ivan Zuluaga', 'Zuluaga\'s']
fico_wordKeys= ['Federico Gutierrez', 'federico','FedericoGutierrez', 'Fico', 'Federico', 'Federico\'s']

# FUNCION LAMBDA PARA IDENTIFICAR CANTIDAD DE MENCIONES #

def identificar_candidato(tweet, refs):
    flag = 0
    for ref in refs:
        if tweet.find(ref) != -1:
            flag = 1
    return flag

df['Petro'] = df['Tweet'].apply(lambda x: identificar_candidato(x, petro_wordKeys))
df['Zuluaga'] = df['Tweet'].apply(lambda x: identificar_candidato(x, zuluaga_wordKeys))
df['Fico'] = df['Tweet'].apply(lambda x: identificar_candidato(x, fico_wordKeys))

df.head(60)
```

```

import nltk
from nltk.corpus import stopwords
from textblob import Word, TextBlob

nltk.download('stopwords')
nltk.download('wordnet')
stop_words = stopwords.words('spanish')
palabras_personalizadas = ['RT', '#DebatePresidencial']
```

```
# PRE-PROCESADO DE LOS TWEETS #

def preprocesamiento(tweet, stopWords):
    processed_tweet = tweet
    processed_tweet.replace('[^\w*\s*]', '')
    processed_tweet = " ".join(palabra for palabra in processed_tweet.split() if palabra not in stop_words)
    processed_tweet = " ".join(palabra for palabra in processed_tweet.split() if palabra not in palabras_personalizadas)
    processed_tweet = " ".join(palabra for palabra in processed_tweet.split() if palabra.isalnum())
    processed_tweet = " ".join(Word(palabra).lemmatize() for palabra in processed_tweet.split())
    return(processed_tweet)

df['Tweet Procesado'] = df['Tweet'].apply(lambda x: preprocesamiento(x, palabras_personalizadas))
df.head()

print('Ejemplo: \n', df['Tweet'][0])
print('Ejemplo limpio y lematizado:\n', df['Tweet Procesado'][0])
```

```
# An√°lisis de sentimientos
# Polaridad 1 -> Positivo, -1 -> Negativo
# Subjetividad 0 -> Objetivo, 1 -> Subjetivo
df['polaridad'] = df['Tweet Procesado'].apply(lambda x: TextBlob(x).sentiment[0])
df['subjetividad'] = df['Tweet Procesado'].apply(lambda x: TextBlob(x).sentiment[1])
df[['Tweet Procesado', 'Petro', 'Zuluaga', 'Fico', 'polaridad', 'subjetividad']].head(60)
```

```
print('Promedio de menciones tweets Gustavo Petro:',round(df['Petro'].mean()*100),'%')
print('Promedio de menciones tweets Oscar Zuluaga:',round(df['Zuluaga'].mean()*100),'%')
print('Promedio de menciones tweets Federico Gutierrez:',round(df['Fico'].mean()*100),'%')
```
